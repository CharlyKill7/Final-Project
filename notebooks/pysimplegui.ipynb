```python
import PySimpleGUI as sg

# Define los parámetros de la ventana
window_title = "Ejemplo de ventana"
window_size = (400, 300)
window_location = (None, None)
resizable = True
alpha_channel = 1.0
disable_close = False
disable_minimize = False
disable_maximize_button = False
finalize = True

# Define el diseño de la ventana
layout = [
    [sg.Text("¡Hola, mundo!")]
]

# Crea la ventana con los parámetros definidos
window = sg.Window(
    title=window_title,
    layout=layout,
    size=window_size,
    location=window_location,
    resizable=resizable,
    alpha_channel=alpha_channel,
    disable_close=disable_close,
    disable_minimize=disable_minimize,
    disable_maximize_button=disable_maximize_button,
    finalize=finalize
)

# Lee los eventos de la ventana
while True:
    event, values = window.read()
    if event == sg.WINDOW_CLOSED:
        break

# Cierra la ventana y finaliza la aplicación
window.close()
```

```python
import PySimpleGUI as sg

sg.theme('DarkPurple6')

# Define los parámetros de la ventana
window_title = "LUNA"
window_size = (400, 300)
window_location = (865, 390)
resizable = True
alpha_channel = 0.9
disable_close = False
disable_minimize = False
finalize = True

# Define el diseño de la ventana
layout = [
    [sg.Text("¡Hola, mundo!")],
    #[sg.Image(filename=r"img\luna.png", background_color='white', size=(300, 300))]
]

# Crea la ventana con los parámetros definidos
window = sg.Window(
    title=window_title,
    layout=layout,
    size=window_size,
    location=window_location,
    resizable=resizable,
    alpha_channel=alpha_channel,
    disable_close=disable_close,
    disable_minimize=disable_minimize,
    finalize=finalize,
)

window.TKroot.attributes("-topmost", True)

# Lee los eventos de la ventana
while True:
    event, values = window.read()
    if event == sg.WIN_CLOSED:
        break

# Cierra la ventana y termina el programa
window.close()
```
